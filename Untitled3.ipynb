```python
#Exercise 1 (repeated this to keep all the code in one block)

from numpy import random
my_list=[]
sum=0.0

my_list=[i for i in range(1,14) for j in range(1,5)]

for i in my_list:
    sum=sum+i

def run_function(): #function which has both deal and play functions defined within
    
    def deal(): #Exercise 2
        random.shuffle(my_list)
        print(my_list)
        
        player_a=tuple(my_list[:int(len(my_list)/2)])
        
        player_b=tuple(my_list[int(len(my_list)/2):])

        return player_a,player_b

    tuple_a,tuple_b=deal()

    sum_a=0
    sum_b=0
    count=0
    return_list=[]
    
    number_a=len(tuple_a)
    for i in tuple_a:
        sum_a=sum_a+i
    print('no of cards with player a is {} and total initial value is {}'.format(number_a,sum_a))

    number_b=len(tuple_b)
    for i in tuple_b:
        sum_b=sum_b+i
    print('no of cards with player b is {} and total initial value is {}'.format(number_b,sum_b))
    list_a=list(tuple_a)
    list_a.reverse()   # converted to list to use the pop function as tuples are immutable and reversed because pop eliminates the last element of a list
    
    list_b=list(tuple_b)
    list_b.reverse()
    
    
# STEP 3.2 and 3.3 excuted below


    def play(): #game of ADDITION WAR is being played within this
        total_a=0
        total_b=0
        
        round=1
        discard_a=[] #this will have the final cards in hand of player A after a game is complete
        int_a=[] # if a game round is TIE , this is an intermediate array where the cards get stored
        discard_b=[] # this will have final cards in hand of player B after a game is complete
        int_b=[]
        while list_a and list_b: #terminating condition specified here
            print('round number is {}'.format(round))
            flag=0
            a_1=list_a.pop() #select card 1 for player A
            a_2=list_a.pop() #select card 2 for player A
            b_1=list_b.pop() #similar for player B
            b_2=list_b.pop()
            value_a=a_1+a_2 #value of two card points computed
            value_b=b_1+b_2
            

            if value_a==value_b: #same card points for both players 
                int_a.append(a_1)
                int_a.append(a_2)
                int_b.append(b_1)
                int_b.append(b_2)
                

            elif value_a>value_b: #player A has a higher card point than player B
                total_int_list=int_a + int_b
                int_a.clear()
                int_b.clear()
                discard_a=discard_a+total_int_list
                discard_a.append(a_1) #values get added to player A's final cards array 
                discard_a.append(a_2)
                discard_a.append(b_1)
                discard_a.append(b_2)

            elif value_a < value_b: #player B has a higher card point than player A
                total_int_list=int_a + int_b
                int_a.clear()
                int_b.clear()
                discard_b=discard_b+total_int_list
                discard_b.append(a_1) #values get added to player B's final cards array
                discard_b.append(a_2)
                discard_b.append(b_1)
                discard_b.append(b_2)
            print(" card points with player A after round {} is {} \n and card points with player B after round {} is {}\n".format(round,discard_a,round,discard_b))
            round+=1

        print('final card point array of player A is : {}'.format(discard_a))
        print('final card point array of player B is :{}'.format(discard_b))
       

        for i in discard_a:
            total_a=total_a+ i #computing total values to make the decision of who won 
        for j in discard_b:
            total_b=total_b+ j
        if (total_a==total_b):
            print('TIE GAME!!!!')
            
        elif total_a>total_b and sum_a<sum_b: #condition included to get an idea about whether player won even when its initial card point array had a lower value than other player
            print('A is the winner as standing is {}'.format(total_a))
            count=1
            flag=0
            return total_a,count,flag,sum_a
        
        elif total_a>total_b:
            print('A is the winner as standing is {}'.format(total_a))
            count=0
            flag=0
            return total_a,count,flag,sum_a
        
        elif total_a<total_b and sum_a>sum_b:
            print('B is the winner as standing is {}'.format(total_b))
            count=0
            flag=1
            return total_b,count,flag,sum_b
        
        else:
            print('B is the winner as standing is {}'.format(total_b))
            count=0
            flag=0
            return total_b,count,flag,sum_b
        
    winner,cnt,flg,init_value = play()
    #return_list is the list that run_function() returns
    return_list.append(winner)
    return_list.append(cnt)
    return_list.append(flg)
    return_list.append(init_value)
    
    return return_list
    
        


        
```

```python
import numpy

win_array=[] #keep a track of winning values 
lose_array=[] # keep a track of loser's card point values
initial_value_array=[] #keep a track of initial card point values of players who won
final_count=0 #this will give us the answer to step 4.2

for i in range(1,21): #playing the game of addition war 20 times
    
    print('\n GAME NO: {}'.format(i))
    vats_list=run_function()
    
    winner_catch=vats_list[0]
    loser_value=364 - winner_catch
    count_catch=vats_list[1]
    flag_catch=vats_list[2]
    if (count_catch==1 and flag_catch==0) or (count_catch==0 and flag_catch==1): #this is necessary to know whether the player who won had a lower initial value or not 
        final_count+=1
    initial_value=vats_list[3]
    
    win_array.append(winner_catch)
    lose_array.append(loser_value)
    initial_value_array.append(initial_value)
    

np_win=numpy.array(win_array) #convert to numpy array to carry out following operations

#following two numpy arrays are used for calculation of linear correlation coefficient
np_initial=numpy.array(initial_value_array)
np_final=numpy.array(win_array)

np_min=numpy.min(np_win) #minimum value of array which has all winner card point values
np_max=numpy.max(np_win) #maximum value of array which has all winner card point values
np_mean=numpy.mean(np_win) #mean value of the same array stated above

np_corr=numpy.corrcoef(np_initial,np_final) #calculation of correlation coefficient

print('\narray which includes all winning card point values : {}'.format(win_array))
print('\narray which includes initial card point value for corresponding winners: {}'.format(initial_value_array))
print('\narray for the loser values :{}'.format(lose_array))
print('\ncount of games where player who had lower initial value won :{}'.format(final_count))
print('\nmin value out of all winner values :{}'.format(np_min))
print('\nmax value out of all winner values :{}'.format(np_max))
print('\nmean value of all the winner values :{}'.format(np_mean))
print('\ncorrelation between initial and final values of the winners:\n {}'.format(np_corr))
```
